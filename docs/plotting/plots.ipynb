```python
import matplotlib.pyplot as plt

input_sizes = [1, 100, 10000, 100000, 1000000, 10000000]

times = {
  "stl"               : [0, 0,  0, 4,  48,  561 ],
  "bitonic_plain"     : [0, 0,  3, 27, 287, 5452],
  "bitonic_insertion" : [0, 0,  2, 23, 263, 5039],
  "bitonic_omp"       : [0, 13, 3, 10, 78,  1362],
  "bitonic_threads"   : [0, 0,  6, 13, 70,  1192],
}

plt.figure(figsize=(10, 6))
for name, time_values in times.items():
  plt.plot(input_sizes, time_values, label=name, marker='o')
plt.title("Czasy algorytmów sortujących lokalnie")
plt.xlabel("Rozmiar danych")
plt.ylabel("Czas (ms)")
plt.legend()
plt.grid()
plt.savefig("czasy_lokalne.jpg")
plt.show()
```
